```python
import pandas as pd
```

```python
df1 = pd.read_csv('threads.csv')
df2 = pd.read_csv('thread_comments.csv')
```

```python
df1.rename(columns={'id': 'thread_id'}, inplace=True)
```

```python
df3 = pd.merge(df1, df2, on = 'thread_id')
df3.set_index('id', inplace = True)
```

```python
df3.head()
```

```python
df_sorted = df3.sort_values(by='created_at_y')
```

```python
df_sorted.head(10)
```

```python
df_sorted.drop(['user_id_x', 'anonymous', 'created_at_x',	'updated_at_x',	'deleted_at_x',	'trending',	'pinned', 'pinning_order',
          'user_id_y', 'is_retrieved',	'retrieved_at',	'is_answer',	'created_at_y',	'updated_at_y',	'deleted_at_y'], axis=1, inplace=True)
```

```python
df_sorted.rename(columns={'content_x': 'Question', 'content_y': 'Answer'}, inplace=True)
```

```python
df_sorted.head()
```

```python
cell_value = df_sorted.loc[1, 'Answer']
print(cell_value)
```

```python

# Define the mapping dictionary
category_mapping = {
    1: 'General Health',
    2: 'Child Health',
    3: 'Mental Health',
    4: 'Baby Care',
    5: 'Infections',
    6: 'Nutrition',
    7: 'Drugs',
    8: 'Skin Care',
    9: 'Fitness',
    10: 'Dental Health',
    11: 'Sexual Health',
    12: "Women's Health",
    13: 'Pregnancy',
    14: 'Eye Health',
    15: "Men's Health",
    16: 'COVID-19'
}

# Apply the mapping
df_sorted['category_name'] = df_sorted['category_id'].map(category_mapping)

# Display the first few rows
df_sorted.head()
```

```python
df_sorted.dropna(inplace=True)
df_sorted.drop_duplicates(inplace=True)
df_sorted['Question'] = df_sorted['Question'].str.replace(r'[^\w\s]', '', regex=True)
df_sorted['Answer'] = df_sorted['Answer'].str.replace(r'[^\w\s]', '', regex=True)
```

```python
from transformers import AutoTokenizer

# Load pre-trained tokenizer (change 'bert-base-uncased' to your specific model)
tokenizer = AutoTokenizer.from_pretrained('gpt2')

# Tokenize and count tokens
df_sorted['Question_token_count'] = df_sorted['Question'].apply(lambda x: len(tokenizer.tokenize(x)))
```

```python
# Tokenize and count tokens
df_sorted['Answer_token_count'] = df_sorted['Answer'].apply(lambda x: len(tokenizer.tokenize(x)))
```

```python
df_sorted['Question_text_length'] = df_sorted['Question'].apply(len)
df_sorted['Answer_text_length'] = df_sorted['Answer'].apply(len)
```

```python
# Count empty strings
empty_strings = df_sorted['Answer'].str.strip().eq('').sum()
print(f"Empty strings in 'Answer': {empty_strings}")

# Count NaN values
nan_values = df_sorted['Answer'].isna().sum()
print(f"NaN values in 'Answer': {nan_values}")
```

```python
# Remove rows where 'Answer' is NaN or empty
df_sorted = df_sorted[df_sorted['Answer'].notna() & df_sorted['Answer'].str.strip().astype(bool)]
```

```python
from langdetect import detect_langs, LangDetectException

def detect_language(text):
    if pd.isnull(text) or not text.strip():
        return None, None
    try:
        lang_probs = detect_langs(text)
        if lang_probs:
            lang = lang_probs[0]
            return lang.lang, lang.prob
        else:
            return None, None
    except LangDetectException:
        return None, None

# Apply to 'Question' column
df_sorted[['Question_language', 'Question_language_probability']] = df_sorted['Question'].apply(
    lambda x: pd.Series(detect_language(x), index=['Question_language', 'Question_language_probability'])
)

# Apply to 'Answer' column
df_sorted[['Answer_language', 'Answer_language_probability']] = df_sorted['Answer'].apply(
    lambda x: pd.Series(detect_language(x), index=['Answer_language', 'Answer_language_probability'])
)
```

```python
desired_order = [
    'thread_id',
    'category_id',
    'category_name',
    'Question',
    'Question_token_count',
    'Question_text_length',
    'Question_language_probability',
    'Answer',
    'Answer_token_count',
    'Answer_text_length',
    'Answer_language_probability'
]


# Assuming your DataFrame is named df
df_sorted = df_sorted[desired_order]
```
